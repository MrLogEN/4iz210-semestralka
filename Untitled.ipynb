## Data Exploration
### 1. Histogramy

```python
target_hist = df.Outcome.value_counts().plot(kind='bar')
target_hist.set_title("Target variable")
target_hist.set_xlabel('Outcome')
target_hist.set_ylabel('Outcome count')
# df.Outcome.values
```

```python
plt.hist(df.Glucose, rwidth=0.9, bins=40)
plt.xlabel("Glucose level")
plt.ylabel("Frequency") 
plt.title("Glucose")
plt.show()
```

```python
plt.hist(df.BloodPressure, rwidth=0.9, bins=40)
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency") 
plt.title("Blood presure")
plt.show()
```

```python
plt.hist(df.Insulin, rwidth=0.9, bins=40)
plt.xlabel("Insulin")
plt.ylabel("Frequency") 
plt.title("Insulin level")
plt.show()
```
